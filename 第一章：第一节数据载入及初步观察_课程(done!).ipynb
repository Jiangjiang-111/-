```python
df = pd.read_csv("/content/drive/MyDrive/hands-on-data-analysis-master/第一单元项目集合/train.csv",
                 names=['乘客ID', '是否幸存', '仓位等级', '姓名', '性别', '年龄', '兄弟姐妹个数', '父母子女个数', '船票信息', '票价', '客舱', '登船港口'],
                 index_col="乘客ID",header=0)
df.head()
# 读取目标文件，将表头依次修改，将索引修改
# names：指定每一列的名称，如果文件中不包含header的行，应显性表示header=None。
# index_col:是read_csv中的一个参数。用来指定表格的索引值。
```

关于index_col

在默认为None的时候，pandas会自动将第一列作为索引，并额外添加一列。如下表：
1663072531524.png

一般来说，我们会使用index_col=0，直接将第一列作为索引，不额外添加列。

在这里我们按照题目要求要把乘客ID作为索引

```python
#法二
#通过.rename进行修改
df = pd.read_csv("/content/drive/MyDrive/hands-on-data-analysis-master/第一单元项目集合/train.csv")
df.rename(columns={'PassengerId':'乘客ID','Survived':'是否幸存','Pclass':'乘客等级(1/2/3等舱位)',
            'Name':'乘客姓名','Sex':'性别','Age':'年龄','SibSp':'堂兄弟/妹个数','Parch':'父母与小孩个数',
            'Ticket':'船票信息','Fare':'票价','Cabin':'客舱','Embarked':'登船港口'},inplace = True)
df1 = df.set_index('乘客ID',inplace = True)
df.head()
```
