#### 1.5.1 ‰ªªÂä°‰∏ÄÔºö Êàë‰ª¨‰ª•"Age"‰∏∫Á≠õÈÄâÊù°‰ª∂ÔºåÊòæÁ§∫Âπ¥ÈæÑÂú®10Â≤Å‰ª•‰∏ãÁöÑ‰πòÂÆ¢‰ø°ÊÅØ„ÄÇ

**reset_index()** ÈáçÁΩÆÁ¥¢ÂºïÔºåÂ∞ÜÊóßÁ¥¢ÂºïÊ∑ªÂä†‰∏∫ÂàóÔºåÂπ∂‰ΩøÁî®Êñ∞ÁöÑÈ°∫Â∫èÁ¥¢ÂºïÔºåÂèØ‰ª•‰ΩøÁî®dropÂèÇÊï∞Êù•ÈÅøÂÖçÂ∞ÜÊóßÁ¥¢ÂºïÊ∑ªÂä†‰∏∫Âàó


ÂáΩÊï∞‰∏ªË¶ÅÊúâ‰ª•‰∏ãÂá†‰∏™ÂèÇÊï∞Ôºö
