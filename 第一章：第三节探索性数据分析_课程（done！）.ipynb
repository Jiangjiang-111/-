#### 1.6.1 任务一：利用Pandas对示例数据进行排序，要求升序

```python
# 具体请看《利用Python进行数据分析》第五章 排序和排名 部分

#自己构建一个都为数字的DataFrame数据

df1 = pd.DataFrame(np.arange(8).reshape((2,4)),index=[2,1],columns=['d', 'a', 'b', 'c'])
df1

```

【代码解析】

pd.DataFrame() ：创建一个DataFrame对象 

np.arange(8).reshape((2, 4)) : 生成一个二维数组（2*4）,第一列：0，1，2，3 第二列：4，5，6，7

index=['2, 1] ：DataFrame 对象的索引列

columns=['d', 'a', 'b', 'c'] ：DataFrame 对象的索引行

1.6.1 升序，排列

**reindex** 重新索引

`.reindex([ ],method="")`


*   默认为索引行，需要索引列则使用column关键字 columns=[ ]

*   若某个索引值不存在，就引入缺失值，可以通过fill_value指定；插值method选项,有参数ffill，bfill

> 对于DataFrame

**sort_index** 对索引进行排列

*   axis= ,可以根据任意一个轴上的索引进行排序
*   ascending= ,默认为升序，当值为False时可以降序排列
*   ~by= ，根据一个或多个列的值进行排序~【注意：pandas模块更新，对有些方法进行重写】

**sort_values** 对值进行排列
*   by= ，根据一个或多个列的值进行排序


> 对于Series

**order** 进行排列


【问题】：大多数时候我们都是想根据列的值来排序,所以将你构建的DataFrame中的数据根据某一列，升序排列

```python
#回答代码
df1 = df1.sort_values(by='d',ascending=False)
df1
```

【思考】通过书本你能说出Pandas对DataFrame数据的其他排序方式吗？

【总结】下面将不同的排序方式做一个总结

1.让行索引升序排序

```python
#代码
#这个方法不是排序，属于重新索引了
df1 = df1.reindex([1,2])
df1
```

```python
df1 = df1.sort_index( )#默认行索引，默认升序
df1
```

2.让列索引升序排序


```python
#代码
df1 = df1.sort_index(axis=1)
df1
```

3.让列索引降序排序

```python
#代码
df1 = df1.sort_index(axis=1,ascending=False)
df1
```

4.让任选两列数据同时降序排序

```python
#代码
df1 = df1.sort_values(by=["a","d"],ascending=False)
df1
```

1. 使用加号、减号等运算符

进行算术运算时，不重叠的索引处引入了NA值（NaN）

当在DataFrame和Series的计算中，默认匹配列，沿着行向下广播

2. 算术方法 

可以使用在DataFrame和Series的计算中

| 方法 | 说明 |
| ---- | ---- |
| add | + |
| sub | - |
| div | % |
| mul | * |

*  axis= ,可以指定匹配 行 或者 列 
*  fill_value= ,可以指定填充值

```python
# 具体请看《利用Python进行数据分析》第五章 算术运算与数据对齐 部分

#自己构建两个都为数字的DataFrame数据
frame1_a = pd.DataFrame(np.arange(9.).reshape(3,3),columns=["a","b","c"],index=["one","two","three"])
frame1_b = pd.DataFrame(np.arange(12.).reshape(4,3),columns=["a","e","c"],index=["first","one","two","second"])
frame1_a.head()                       
```
